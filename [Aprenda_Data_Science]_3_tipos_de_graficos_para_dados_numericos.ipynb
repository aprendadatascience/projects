```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
# gera dados para regressão
X1, y1 = make_regression(
    n_samples=200,
    n_features=4,
    n_informative=4,
    noise = 10,
    random_state=25
    )
print(X1[:5])
```

## Scatter Plot

```python
# Gerando polinômio interpolado para variáveis de indice 0
m, b = np.polyfit(X1[:, 0], y1, 1)
best_fit = m*X1 + b# plotting
plt.subplots(figsize=(8, 5))
plt.scatter(X1[:, 0], y1, color="blue")
plt.plot(X1, best_fit, color="red")
plt.title("Relacionamento da variável 0 com o Y")
plt.xlabel("Feature no indice 0")
plt.ylabel("Variável Alvo (y)")
plt.show()
```

```python
# Gerando polinômio interpolado para variáveis de indice 1
m, b = np.polyfit(X1[:, 1], y1, 1)
best_fit = m*X1 + b# plotting
plt.subplots(figsize=(8, 5))
plt.scatter(X1[:, 1], y1, color="blue")
plt.plot(X1, best_fit, color="red")
plt.title("Relacionamento da variável 0 com o Y")
plt.xlabel("Feature no indice 1")
plt.ylabel("Variável Alvo (y)")
plt.show()
```

```python
# Gerando polinômio interpolado para variáveis de indice 2
m, b = np.polyfit(X1[:, 2], y1, 1)
best_fit = m*X1 + b# plotting
plt.subplots(figsize=(8, 5))
plt.scatter(X1[:, 2], y1, color="blue")
plt.plot(X1, best_fit, color="red")
plt.title("Relacionamento da variável 2 com o Y")
plt.xlabel("Feature no indice 2")
plt.ylabel("Variável Alvo (y)")
plt.show()
```

## Histograma
